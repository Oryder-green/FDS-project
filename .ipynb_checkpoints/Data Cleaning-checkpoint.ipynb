```python
URL = "https://www.bts.gov/browse-statistical-products-and-data/bts-publications/airline-service-quality-performance-234-time"
r = requests.get(URL)
print(r.content)
```

```python
soup = BeautifulSoup(r.content, 'html5lib') # If this line causes an error, run 'pip install html5lib' or install html5lib
print(soup.prettify())
```

```python
# find all a href links available on this page
file_urls = soup.findAll('a', href=True)

links = []
years = [2018, 2017]

for year in years:
    for link in file_urls:
        spec_link = link['href']
        if spec_link.endswith(f'{year}.zip') or spec_link.endswith(f'{year}.ZIP'):
            links.append(spec_link)

```

```python
links
```

```python
### remove first 10 file links and final file link as these belong to data from 2018 and 2016

del links[0:10]
del links[-1:]
```

```python
links
```

```python
# download all the files corresponding to those href links, unzip, and 
# make into a pandas dataframe. Note: files are in .asc format. 
```

```python
full_df = pd.DataFrame()
url= 'https://www.bts.gov'

for link in links:
    #open url
    resp = urlopen(url+link)
    #read zipfile
    zipfile = ZipFile(BytesIO(resp.read()))
    #get the asc file name.
    fname = zipfile.namelist()[0]
    #convert to pandas dataframe
    df = pd.read_csv(zipfile.open(fname), delimiter = '|', dtype=object, header=None)
    zipfile.close()
    
    full_df = pd.concat([full_df, df], ignore_index=True, sort=False)

    print(f'{fname} Completed')
```

```python
### Scrape historical weather data from Weather Underground. 

def scrape_station(station):
    current_date = datetime(year=2014, month=7, day=1)
    end_date = datetime(year=2015, month=7, day=1)

    # Make sure a directory exists for the station web pages
    os.mkdir(station)

    # Use .format(station, YYYY, M, D)
    lookup_URL = 'http://www.wunderground.com/history/airport/{}/{}/{}/{}/DailyHistory.html'

    while current_date != end_date:

        if current_date.day == 1:
            print(current_date)

        formatted_lookup_URL = lookup_URL.format(station,
                                                 current_date.year,
                                                 current_date.month,
                                                 current_date.day)
        html = urlopen(formatted_lookup_URL).read().decode('utf-8')

        out_file_name = '{}/{}-{}-{}.html'.format(station, current_date.year,
                                                  current_date.month,
                                                  current_date.day)

        with open(out_file_name, 'w') as out_file:
            out_file.write(html)

        current_date += timedelta(days=1)
```

```python
# Scrape the stations that appear in the flights dataset
for station in ['KCLT', 'KCQT', 'KHOU', 'KIND', 'KJAX',
                'KMDW', 'KNYC', 'KPHL', 'KPHX', 'KSEA']:
    scrape_station(station)
```
